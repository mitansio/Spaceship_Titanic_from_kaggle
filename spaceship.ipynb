```python
train[['GroupID', 'PersonID']] = train['PassengerId'].str.split('_', expand=True)
test[['GroupID', 'PersonID']] = test['PassengerId'].str.split('_', expand=True)
```

```python
import pandas as pd

# Function to fill missing values group-wise
def fill_missing_groupwise(df):
    groups = df['GroupID'].unique()

    for val in groups:
        group_data = df[df['GroupID'] == val]

        for column in group_data.columns:
            if group_data[column].isnull().sum() > 0:

                # Age  group mean (only if mean exists)
                if column == 'Age':
                    mean_age = group_data[column].mean()
                    if not pd.isna(mean_age):
                        df.loc[df['GroupID'] == val, column] = (
                            group_data[column].fillna(int(mean_age))
                        )

                # Numeric columns  group median
                elif group_data[column].dtype in ['int64', 'float64']:
                    median_val = group_data[column].median()
                    if not pd.isna(median_val):
                        df.loc[df['GroupID'] == val, column] = (
                            group_data[column].fillna(median_val)
                        )
              

                # Categorical columns  group mode
                else:
                    mode_val = group_data[column].mode()
                    if not mode_val.empty:
                        df.loc[df['GroupID'] == val, column] = (
                            group_data[column].fillna(mode_val.iloc[0])
                        )
    return df

# Apply to train and test
train = fill_missing_groupwise(train)
test  = fill_missing_groupwise(test)
```

```python
columns=train.columns.tolist()
for col in columns:
    if train[col].isnull().sum():
        if col == 'Age' :
            train[col].fillna(int(train[col].mean()), inplace=True)
            test[col].fillna(int(test[col].mean()), inplace=True)
        elif train[col].dtype in ['int64', 'float64']:
            train[col].fillna(train[col].median(),inplace=True)
            test[col].fillna(test[col].median(),inplace=True)
        else:
            if train[col].dtype in ['object']:
                train[col].fillna(train[col].mode()[0],inplace=True)
                test[col].fillna(test[col].mode()[0],inplace=True)

        
```

### 3 Feature Engineering

#### 3.1 feature creation

```python
train['TotalSpent'] = train.RoomService + train.FoodCourt + train.ShoppingMall + train.Spa + train.VRDeck
test['TotalSpent'] = test.RoomService + test.FoodCourt + test.ShoppingMall + test.Spa + test.VRDeck
```

#### 3.2 Encoding

```python
test_id=test.PassengerId
train.drop(['PassengerId','Name','GroupID','PersonID'], axis=1, inplace=True)
test.drop(['PassengerId','Name','GroupID'], axis=1, inplace=True)
```

```python
cat_col=train.select_dtypes(include=['object']).columns.tolist()
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()


for col in cat_col:
    train[col] = le.fit_transform(train[col].astype(str))
    test[col] = le.fit_transform(test[col].astype(str))
```

```python
boolean = train.select_dtypes(include=['bool']).columns.tolist()
boolean_= test.select_dtypes(include=['bool']).columns.tolist()
for col in boolean:
    train[col] = train[col].map({True:1, False:0})
   
for col in boolean_:
     test[col] = test[col].map({True:1, False:0})
```

### 4 Data Preparation

```python
from sklearn.model_selection import train_test_split
features=['FoodCourt',
 'VRDeck',
 'Spa',
 'ShoppingMall',
 'Cabin',
 'TotalSpent',
 'RoomService',
 'CryoSleep',
 'Age',
 'HomePlanet']
y=train['Transported']

x=train[features]
x_test=test[features]
x_train,x_val,y_train,y_val =train_test_split(x,y, test_size=0.2, shuffle=True, random_state=42)
```

### 5 Model

#### 5.1 Baseline Model

```python
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Split is already done
# x_train, x_val, y_train, y_val

# Determine majority class in training set
majority_class = y_train.mode()[0]
print("Majority class:", majority_class)

# Predict baseline on validation set
y_pred_baseline = np.full_like(y_val, fill_value=majority_class)


# Evaluate baseline
accuracy = accuracy_score(y_val, y_pred_baseline)
f1 = f1_score(y_val, y_pred_baseline)

print("Baseline Accuracy:", accuracy)
print("Baseline F1-score:", f1)
print("\nClassification Report:\n", classification_report(y_val, y_pred_baseline))
```

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score

models = {
    "GradientBoosting": GradientBoostingClassifier(
        n_estimators=1000, learning_rate=0.05, max_depth=5, random_state=42
    ),
    "LogisticRegression": LogisticRegression(max_iter=1000, random_state=42),
    "RandomForest": RandomForestClassifier(
        n_estimators=500, max_depth=None, min_samples_leaf=5, max_features='sqrt', random_state=42, n_jobs=-1
    ),
    "XGBClassifier": XGBClassifier(
        n_estimators=300, learning_rate=0.05, max_depth=4, subsample=0.8,
        colsample_bytree=0.8, reg_alpha=0.5, reg_lambda=1.0, random_state=42
    )
}

for name, model in models.items():
    # Fit on training set
    model.fit(x_train, y_train)
    
    # Predict on validation set
    y_pred = model.predict(x_val)
    
    # Evaluate
    acc = accuracy_score(y_val, y_pred)
    f1 = f1_score(y_val, y_pred)
    
    print(f"{name} Accuracy: {acc:.3f}, F1-score: {f1:.3f}")
```
